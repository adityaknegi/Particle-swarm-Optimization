```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
P = 100; 3 #population
C1=0.5;
C2=0.5;
w=0.9;
```

```python
x =np.random.rand(1,50)*2*np.pi
x=x.tolist()[0]
v=np.random.rand(1,50)
v=v.tolist()[0]
```

```python
#personal best
pbest =x # intialized x[i] as pbest for each swarm i 
```

```python
def fun(x):
    return np.sin(x)
```

```python
x=list(zip(x,[fun(i) for i in pbest],pbest))

x[:2]
```

```python
x.sort(key=lambda t :t[1]) # sort with respect to value
gbest = x[49][2] # x with max fun value
gbest_value=x[49][1]
gbest_value
```

```python
g_pre_best =gbest
count=0
max_count=3000
j=0
```

```python

def plot(x):
    y =np.linspace(0,2*np.pi)
    plt.plot(y,fun(y))
    #unzip 
    pos ,fun_value, pos_best = map(list,zip(*x))
    plt.scatter(pos,fun(pos))
    plt.xlabel('Genration {}'.format(j))
    
    
```

```python
plot(x)
```

```python
while  1:
    if(j%300==0 or j<20):
        plot(x)
    
    count=1+count
    w=0.9 -(count*0.5)/max_count
    pos,fun_val ,pbest =map(list,zip(*x))
    pos_new =[]
    for i in range(len(v)):
        #update velocity
        v[i]=w*v[i]+C1*(np.random.rand(1)*(pbest[i]-pos[i]))+C2*(np.random.rand(1)*(gbest-pos[i]))
        pos[i]=pos[i]+v[i][0]
        if(fun(pos[i])>fun_val[i]):
            pbest[i]=pos[i]
            fun_val[i]=fun(pbest[i])
    
    x=list(zip(pos,fun_val,pbest))
    x.sort(key=lambda t : t[1])
    gbest =x[49][2]
    gbest_value=x[49][1]
    if count >max_count:
        break;
    else:
        g_pre_best=gbest
    j=j+1
    plt.show()
    
print('gbest={} , gbest_value={}'.format(gbest,gbest_value))
    

        
            
```
